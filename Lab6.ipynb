```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
```

```python
p = pd.read_csv("points.csv")
p
```

```python
p.info()
```

```python
p.describe()
```

##Завдання 3

```python
p.var()
```

```python
p.corr()
```

##Завдання 5

```python
fig = px.scatter_3d(p, x = "x", y = "y", z = "z")
fig
```

##Завдання 6

```python
fig.update_traces(marker={"size":3})
```

##Завдання 7

```python
fig.update_layout(scene = { "aspectmode": "data" })
```

##Завдання 9

```python
pca = PCA()
pca.fit(p)
pca.get_params()
```

##Завдання 10

```python
p_trans = pd.DataFrame(pca.transform(p), columns = ["c1", "c2", "c3"])
```

##Завдання 11

```python
p_trans.describe()
```

```python
p_trans.var()
```

```python
p_trans.corr()
```

##Завдання 12

```python
fig_trans = px.scatter_3d(p_trans, x = "c1", y = "c2", z = "c3")
fig_trans.update_traces(marker={ "size": 3 })
fig_trans.update_layout(scene = { "aspectmode": "data" })
fig_trans
```

##Завдання 13

```python
p_trans_dimred = p_trans.copy()
p_trans_dimred["c3"] = 0
p_trans_dimred
```

##Завдання 14

```python
fig_trans.add_scatter3d(x = p_trans_dimred.c1, y = p_trans_dimred.c2, z = p_trans_dimred.c3, mode = "markers", marker_size = 3)
```

##Завдання 15

```python
p_dimred = pd.DataFrame(pca.inverse_transform(p_trans_dimred), columns = ["x", "y", "z"])
p_dimred
```

##Завдання 16

```python
fig.add_scatter3d(x = p_dimred.x, y = p_dimred.y, z = p_dimred.z, mode = "markers", marker_size = 3)
```
